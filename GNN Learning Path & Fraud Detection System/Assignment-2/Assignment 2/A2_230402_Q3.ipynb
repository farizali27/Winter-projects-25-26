```python
import numpy as np
```

```python


def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(a):
    return a * (1 - a)

def XOR(X,y,lr,epochs):
    W1 = np.zeros((2,2))
    b1 = np.zeros((1, 2))
    
    W2 = np.zeros(2, 1)
    b2 = np.zeros((1, 1))
    
    for epoch in range(epochs):
        z1 = X @ W1 + b1
        a1 = sigmoid(z1)
        
        z2 = a1 @ W2 + b2
        y_hat = sigmoid(z2)
        
        loss = np.mean((y - y_hat) ** 2)
        d_y_hat = 2 * (y_hat - y) / y.size
        d_z2 = d_y_hat * sigmoid_derivative(y_hat)
        
        dW2 = a1.T @ d_z2
        db2 = np.sum(d_z2, axis=0, keepdims=True)

        d_a1 = d_z2 @ W2.T
        d_z1 = d_a1 * sigmoid_derivative(a1)
        
        dW1 = X.T @ d_z1
        db1 = np.sum(d_z1, axis=0, keepdims=True)
        
        W2 -= lr * dW2
        b2 -= lr * db2
        W1 -= lr * dW1
        b1 -= lr * db1
    
    print("predictions:")
    print((y_hat > 0.5).astype(int))
```
