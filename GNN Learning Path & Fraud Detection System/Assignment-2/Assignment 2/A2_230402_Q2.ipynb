```python
import numpy as np
```

```python
class LinearSVM:
    def __init__(self,learning_rate, lambda_param, n_iters):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
    
    def fit(self,X,y):
        X_train = np.array(X)
        y_train = np.array(y)

        num_cols = X_train.shape[1]
        w = np.zeros(num_cols)
        b = 0
        y_train[y_train == 0] = -1

        epochs = self.n_iters
        lr = self.learning_rate
        lambda_param = self.lambda_param
        for i in range(epochs):
            for xi, yi in zip(X_train, y_train):
                if yi*(np.dot(w,xi)-b) >= 1:
                    w -= lr*2*lambda_param*w
                else:
                    w -= lr*(2*lambda_param*w - yi*xi)
                    b -= lr*yi
        
        self.w=w
        self.b=b

    def predict(self,X):
        w=self.w
        b=self.b
        
        f = np.dot(w,X) - b
        return np.sign(f)

                    
```
